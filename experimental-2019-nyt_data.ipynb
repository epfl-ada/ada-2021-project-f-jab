```python
import bz2
import json
import pandas as pd 
from tqdm import tqdm 
```

```python
def dict_merge(df, instance):
    
    for key, value in instance.items():
        df[key].append(value)
        
    return df
```

```python
# Specifying filename and directory
filename = 'quotes-2019-nytimes'
data_dir = 'Data'
```

```python


filename = 'quotes-2019-nytimes'
data_dir = 'Data'
path_to_file = rf'.\{data_dir}\{filename}.json.bz2'

dict_df = {}
i=0

with bz2.open(path_to_file, 'rb') as s_file:
    for instance in tqdm(s_file):
        instance = json.loads(instance) # loading a sample
       
        if i == 0:
            dict_df = {key: [value] for key, value in instance.items()}
        else:
            dict_df = dict_merge(dict_df, instance)
        
        i += 1
        
df = pd.DataFrame(dict_df)
df.head()
```
